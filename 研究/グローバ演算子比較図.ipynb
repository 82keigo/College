```python
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
n=100
a=0.2
theta=np.arcsin(a)
d=2**n
p=0.99
Nq=list(range(0,1000,1))
Fg=[]
Fq=[]
SLD=[]
for i in range(1000):
    Fg.append(4*Nq[i]**2*p**(2*Nq[i]))
    Fq.append(4*Nq[i]**2*p**Nq[i]\
              +8*Nq[i]**2*(d-1)*(1-p**Nq[i])**2/d**2\
              -8*Nq[i]**2*(1-p**Nq[i])*(\
                                       np.sqrt(\
                                               (d-1)*(d-1+p**Nq[i])*((d-1)*p**Nq[i]+1)\
                                               )\
                                       )/d**2)
    SLD.append(4*Nq[i]**2*p**(2*Nq[i])/(\
                                       2/d+(1-2/d)*p**Nq[i]\
                                       ))
plt.figure(figsize=(5,5))
plt.xlim(0,1000)
plt.ylim(0,25000)
plt.xlabel("Number of Queries")
plt.ylabel("Fisher Information")
plt.plot(Nq,SLD,label="SLD")
plt.plot(Nq,Fq,linestyle="dashed",label="Fq_envelope")
plt.plot(Nq,Fg,label="Fg_envelope")
plt.legend(bbox_to_anchor=(1, 1), loc='upper right', borderaxespad=0, fontsize=12,prop={"family":"MS Gothic"})
plt.show
```
