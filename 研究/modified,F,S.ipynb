```python
plt.rcParams['font.family'] = 'Times New Roman' # font familyの設定
plt.rcParams['mathtext.fontset'] = 'stix' # math fontの設定
plt.rcParams["font.size"] = 5# 全体のフォントサイズが変更されます。
plt.rcParams['xtick.labelsize'] = 10# 軸だけ変更されます。
plt.rcParams['ytick.labelsize'] = 10# 軸だけ変更されます
plt.rcParams['xtick.direction'] = 'in' # x axis in
plt.rcParams['ytick.direction'] = 'in' # y axis in 
plt.rcParams['axes.linewidth'] = 1.0 # axis line width
plt.rcParams['axes.grid'] = True # make grid
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import math
pi=math.pi
theta=pi/4
m=np.arange(1,100,1)
sin=np.sin(2*m*theta)
cos=np.cos(2*m*theta)
p=0.9
d=10000
D=1/d
print(D)
```

```python
F=(16*(m**2)*(p**(2*m))*(sin**2)*(cos**2))/(D+((cos**2-D)*(p**m))*((1-D)+((sin**2)-(1-D))*(p**m)))
S=(16*d*(m**2)*(p**(2*m)))/(2+(d-2)*p**m)
SS=16*m**2*p**m
plt.plot(m,F)
plt.plot(m,S)
plt.plot(m,SS)
```

```python
P=4/((2/p**m)-1)
J11=P*4*m**2
J12=P*(m**2)/p
J21=P*(m**2)/p
J22=P*(m**2)/(4*p**2)
J11sum=J11.cumsum()
J12sum=J12.cumsum()
J21sum=J21.cumsum()
J22sum=J22.cumsum()
J_11=J22sum/(J11sum*J22sum-J12sum*J21sum)
SLD11=(16*m**2)*p**m
SLD11sum=SLD11.cumsum()
SLD_11sum=1/SLD11sum
Ftheta=(16*(m**2)*(p**(2*m))*(sin**2)*(cos**2))/(((cos**2)*p**m)*(1+((sin**2)-1)*p**m))
```

```python
plt.plot(m,SS,color='tomato',linestyle='solid',label='SLD Fisher Information(d→∞)')
plt.plot(m,Ftheta,color='g',linestyle='solid',label='Classical Fisher Information(d→∞)')
plt.legend(bbox_to_anchor=(1, 1), loc='upper right', borderaxespad=0, fontsize=10)
plt.xlabel(r'$m$',fontsize=12)
plt.ylabel(r'$Fisher\;Information$',fontsize=12)
plt.ylim(0,1000)
plt.show()
```
