```python
import numpy as np
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
 
r = 1 # 半径を指定
theta_1_0 = np.linspace(0, 2*np.pi, 20) # θ_1は[0,π/2]の値をとる
theta_2_0 = np.linspace(0, 2*np.pi, 20) # θ_2は[0,π/2]の値をとる
theta_1, theta_2 = np.meshgrid(theta_1_0, theta_2_0) # ２次元配列に変換
x = np.cos(theta_2)*np.sin(theta_1) * r # xの極座標表示
y = np.sin(theta_2)*np.sin(theta_1) * r # yの極座標表示
z = np.cos(theta_1) * r # zの極座標表示
 
fig = plt.figure() # 描画領域を作成
ax = fig.add_subplot(111, projection="3d") # 3Dの軸を作成
ax.plot_surface(x,y,z) # 球を３次元空間に表示
plt.xlim([-1,1])
plt.ylim([1,-1]) # Y軸の表示は初期値とは逆にする
ax.set_zlim([-1,1])
ax.set_xticks(np.arange(1, -1.1, step=-0.5))
ax.set_yticks(np.arange(-1, 1.1, step=0.5))
ax.set_zticks(np.arange(1, -1.1, step=-0.5))
plt.xlabel("X")
plt.ylabel("Y")
ax.set_zlabel("Z")
plt.subplots_adjust(left=0, right=1, bottom=0, top=1)
ax.view_init(elev=40, azim=30)
plt.show()
```
