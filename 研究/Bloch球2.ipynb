```python
import numpy as np
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
 
r = 1 # 
theta_1_0 = np.linspace(0, 2*np.pi, 20) # _1[0,/2]
theta_2_0 = np.linspace(0, 2*np.pi, 20) # _2[0,/2]
theta_1, theta_2 = np.meshgrid(theta_1_0, theta_2_0) # 
x = np.cos(theta_2)*np.sin(theta_1) * r # x
y = np.sin(theta_2)*np.sin(theta_1) * r # y
z = np.cos(theta_1) * r # z
 
fig = plt.figure() # 
ax = fig.add_subplot(111, projection="3d") # 3D
ax.plot_surface(x,y,z) # 
plt.xlim([-1,1])
plt.ylim([1,-1]) # Y
ax.set_zlim([-1,1])
ax.set_xticks(np.arange(1, -1.1, step=-0.5))
ax.set_yticks(np.arange(-1, 1.1, step=0.5))
ax.set_zticks(np.arange(1, -1.1, step=-0.5))
plt.xlabel("X")
plt.ylabel("Y")
ax.set_zlabel("Z")
plt.subplots_adjust(left=0, right=1, bottom=0, top=1)
ax.view_init(elev=40, azim=30)
plt.show()
```
