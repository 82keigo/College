```python
%matplotlib inline
import numpy as np  # NumPy を読み込む
import matplotlib.pyplot as plt # Matplotlib を読み込む
p=0.99
a=1/2
m = np.arange(0,10000,1)  # x座標を-10 から 10 まで 0.1 きざみで取得
J = ((2*m+1)**2/(a*(1-a)))# 二次関数の式
Jsum=J.cumsum()
JS=p**(2*m)*J
JSsum=JS.cumsum()
SQL=(2*m+1)/(a*(1-a))
SQLsum=SQL.cumsum()

plt.plot(m,SQLsum)
ax = plt.gca()
ax.spines['top'].set_color('none')

##

#ax.set_xscale('log')
##
plt.title('single logarithmic plot') 
plt.xlabel('X',fontsize=18)
plt.ylabel('Y',fontsize=18)

plt.grid(which="both") # グリッド表示。"both"はxy軸両方にグリッドを描く。

plt.show()
```
